**3) Realiza un programa que realice el sumatorio de los primeros N números, empezando en 0, y siendo N el número límite que el usuario proporciona por teclado:** 
* Ejemplo de número límite introducido por el usuario: 5
* sumatorio = 0 + 1 + 2 + 3 + 4 (sumamos los 5 primeros números)

````python
#Creamos la variable del sumatorio
N = int(input('introduzca el numero maximo de la suma:'))

#Creamos desde que numero va a empezar la suma
suma = 0

#Hacemos el bucle for con la funcion range
for x in range(N):
    suma += x

#mostramos el resultado en pantalla
print(f"El sumatorio de los primeros {N} números es: {suma}")
````

**4) Realiza un programa que lea letras y cuente con una variable contador las letras "a" que se introducen. Para salir del programa, introducir el carácter ".". Al finalizar mostrar el número de veces que se ha pulsado la letra "a".**

Ejemplo de salida por pantalla del ejercicio:


````python
#Creamos una variable contador
contador_a = 0

#Hacemos el bucle while
while True:
    letra = input('Introduzca una letra (. para salir):')
    
    if letra == ".":
        break
    
    if letra == "a":
        contador_a += 1

#Mostramos el resultado en pantalla
print("El número de veces que se ha pulsado la letra 'a' es:", contador_a)
````

**5) Realiza el sumatorio de los números pares comprendidos entre 1000 y 73 (en este orden, yendo hacia atrás). Implementa esta solución con un bucle for o while:**

````python
#Asigno una variable
suma = 0

#Creo un bucle For que itere desde ele 1000 hasta el 73 de mayor a menor
for numero in range(1000,73,-1):
    if numero % 2 == 0:
        suma += numero
        
#Muestro en pantalla
print ("La suma de todos los numeros pares en el rango asignado es:", suma)
````

**6) Realiza un programa que solicite al usuario un número que representa el total de números que se van a introducir. Posteriormente leer todos esos números, meterlos en una lista y realizar:**
* La ordenación de la lista de menor a mayor
* Indicar el promedio de la lista
* Indicar el valor máximo y el valor mínimo y en que posiciones se encuentran

Ejemplo de salida por pantalla del ejercicio:


````python
#indicamos la cantidad de numeros que debe introducir el usuario
cantidad = int(input('¿Cuantos numeros vas a introducir?:'))

# Crear una lista para almacenar los números ingresados
lista_num =[]

# Leer y almacenar los números en la lista
for i in range(cantidad):
    numero = int(input("Ingrese un valor entero:"))
    lista_num.append(numero)
    

# Ordenar la lista de menor a mayor
numeros_ordenados = sorted(lista_num)

# Calcular el promedio de la lista
promedio = sum(lista_num) / cantidad

# Encontrar el valor máximo y mínimo, y sus posiciones
maximo = max(lista_num)
minimo = min(lista_num)
posicion_maximo = lista_num.index(maximo)
posicion_minimo = lista_num.index(minimo)

# Imprimir los resultados
print ("La lista de los numeros introducidos es:", lista_num)
print ("La lista de los numeros ordenados es:", numeros_ordenados)
print ("El promedio de la lista es:", promedio)
print (f"El valor máximo de la lista es: {maximo} y se encuentra en la posición: {posicion_maximo + 1}")
print (f"El valor mínimo de la lista es: {minimo} y se encuentra en la posición: {posicion_minimo + 1}")
````

**8) Crea un menú con esta forma:**
```python
MENU COMPRA-VENTA
    1. Comprar
    2. Vender
    3. Alquilar
    4. Mostrar listas de datos
    5. Salir
    Introduzca que desea hacer(1-4): 
```
* Si se selecciona la opción 1 preguntaremos al usuario que quiere comprar y después de que el usuario lo haya introducido se guardará en una lista destinada a las compras y se volverá al menú principal.
* Si se selecciona la opción 2 preguntaremos al usuario que quiere vender y después de que el usuario lo haya introducido se guardará en una lista destinada a las ventas y se volverá al menú principal.
* Si se selecciona la opción 3 preguntaremos al usuario que quiere alquilar y después de que el usuario lo haya introducido se guardará en una lista destinada a los alquileres y se volverá al menú principal.
* Si se selecciona la opción 4 se mostrarán todas las listas.
* Si se selecciona la opción 5 saldremos del bucle.

````python
#Creo las lista vacias para poder meter cosas dentro
comprar = []
vender =[]
alquilar = []

#Empiezo a crear el menu interactivo que va a mostrarse en pantalla
print("MENU COMPRA-VENTA")
while(True):
    print(""" 
    1) Comprar
    2) Vender
    3) Alquilar
    4) Mostrar listas de datos
    5) Salir""")
    #Interactua con el usuario y creo los distintos controles de flujo para ir almacenando en las distintas listas las cosas
    opcion = input("Introduzca una opción: ")
    if opcion == '1':
        compra = input('Que necesitas comprar?:')
        comprar.append(compra)
        print ('Agregado a lista Comprar')
    elif opcion == '2':
        vende = input('Que necesitas vender?:')
        vender.append(vende)
        print ('Agregado a lista Vender')
    elif opcion == '3':
        alquila = input('Que necesitas alquilar?:')
        alquilar.append(alquila)
        print ('Agregado a lista Alquilar')
    #muestro en pantalla lo que contienen cada lista
    elif opcion == '4':
        print (comprar)
        print (vender)
        print (alquilar)
    else: 
        print("Hasta pronto!!!")
        break
````

**9) Realiza un programa que realice las siguientes tareas:**
* __Paso 1__) solicite al usuario una contraseña y la guarde en una variable 
* __Paso 2__) pida al usuario introducir la contraseña para loguearse
    * si la contraseña es correcta mostramos "__Login correcto__"
    * si la contraseña no es correcta se le tendrá que mostrar un mensaje de "__Login fallido. Intento 1__", donde 1 será una variable contador que irá incrementandose hasta llegar a __5__. En ese punto __reiniciamos el programa y volveremos al Paso 1__

````python
#Creo variable contraseña
contraseña = input('introduzca su nueva contraseña:')
print ('su contraeña ha sido almacenada.')

#creo variable intentos
intentos = 0
"""a traves de un bucle while voy iterando para comprobar la contraseña, 
si la mete bien el programa para si falla cinco veces se reinicia pidiendo una nueva contraseña"""
while intentos < 5:
    clave = input('Introduzca su contraseña: ')
    if clave == contraseña:
        print('Login correcto')
        break
    else:
        intentos += 1
        if intentos < 5:
            print('Login fallido. Intento', intentos)
        else:
            print('Demasiados intentos. Reiniciando...')
            intentos = 0
            contraseña = input('Introduzca su nueva contraseña: ')
            print('Su contraseña ha sido actualizada.')

print('Hasta pronto')
````
