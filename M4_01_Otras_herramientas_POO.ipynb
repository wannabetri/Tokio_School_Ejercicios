**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
class Vehiculo():#Superclase 

    def __init__ (self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    def repostar(self):#metodo repostar
        print ("Este vehiculo tiene que repostar gasolina")
        
    def __srt__ (self):
        return " > Marca: {}\n > Modelo: {}".format(self.marca, self.modelo )

class VElectrico():#Superclase 

    def __init__ (self, marca, modelo, autonomia):
        self.marca = marca
        self.modelo = modelo
        self.autonomia = autonomia
        
    def repostar(self):#metodo repostar
        print ("Este vehiculo tiene que repostar electricidad")
        
    def __str__(self):#Defino el metodo srt
        return " > Marca: {}\n > Modelo: {}\n > Autonomia: {}".format(self.marca, self.modelo, self.autonomia)

class BicicletaElectrica(VElectrico, Vehiculo):#Clase con herencia multiple

    def __init__(self, marca=None, modelo=None, autonomia=None):
        VElectrico.__init__(self, marca, modelo, autonomia) # Equivalente a: super().__init__(autonomia)        
        Vehiculo.__init__(self, marca, modelo)
        print("Creado objeto de la clase BicicletaElectrica")
        
    def __str__(self):
        return " > Marca: {}\n > Modelo: {}\n > Autonomia: {}".format(self.marca, self.modelo, self.autonomia)
    

class Quad(Vehiculo, VElectrico):#Clase con herencia multiple

    def __init__(self, marca=None, modelo=None, autonomia=None):
        Vehiculo.__init__(self, marca, modelo)
        VElectrico.__init__(self, marca, modelo, autonomia) # Equivalente a: super().__init__(autonomia)        
        print("Creado objeto de la clase Quad")
        
    def __str__(self):
        return " > Marca: {}\n > Modelo: {}\n > Autonomia: {}".format(self.marca, self.modelo, self.autonomia)
        
        
class Repostaje:#Clase adicional para repostaje
    def combustible(self, vehiculo):
        vehiculo.repostar()


        
#Creamos objetos para probar la herencia y el polimorfismo
miBici = BicicletaElectrica(marca="Orbea", modelo="Orca", autonomia=30)      
miQuad = Quad(marca="Ford", modelo="F-150")
        
miCombustible = Repostaje()
miCombustible.combustible(miBici)
print(miBici, end="\n\n")

miCombustible.combustible(miQuad)
print(miQuad, end="\n\n")
        

```

**2) Ejercicio de herencia**
* Crea la superclase Personal_Universitario con un único atributo que sea un diccionario que incluya: "id", "nombre" y "email"
* Crear la clases Oficina, Profesor y Alumno. Todas ellas heredaran de Personal_Universitario y heredaran sus datos
* En el caso de Oficina, debe añadir a su diccionario de datos el dato "Puesto"
* En el caso de Profesor, debe añadir a su diccionario de datos el dato "Especialización"
* En el caso de Alumno, debe añadir a su diccionario de datos el dato "CreditosAprobados" (integer)
* Crear métodos que sirvan para mostrar la información de cada clase
* Crear objetos de todas las clases
* Mostrar la información de esos objetos y comprobar que sale lo que tiene que salir

```python
class Personal_Universitario:#Superclase 
    def __init__(self, id, nombre, email):
        self.informacion = {"id": id, "nombre": nombre, "email": email}
    def __str__(self):
        return f"ID: {self.informacion['id']}\nNombre: {self.informacion['nombre']}\nEmail: {self.informacion['email']}"
           
class Oficina (Personal_Universitario):#Clase heredera de superclase
    def __init__ (self, id, nombre, email, puesto):#Añadimos el dato puesto
        super().__init__(id, nombre, email)
        self.informacion["puesto"] = puesto
    
    def mostrar_informacion(self):#Metodo para mostrar informacion
        print("Información de la Oficina:")
        print(f"ID: {self.informacion['id']}")
        print(f"Nombre: {self.informacion['nombre']}")
        print(f"Email: {self.informacion['email']}")
        print(f"Puesto: {self.informacion['puesto']}")
       
    def __str__(self):#Clase heredera de superclase
        return f"ID: {self.informacion['id']}\nNombre: {self.informacion['nombre']}\nEmail: {self.informacion['email']}\nPuesto: {self.informacion['puesto']}"
        
class Profesor (Personal_Universitario):#Clase heredera de superclase
    def __init__ (self, id, nombre, email, especializacion):#Añadimos el dato especializacion
        super().__init__(id, nombre, email)
        self.informacion["especializacion"] = especializacion
    
    def mostrar_informacion(self):#Metodo para mostrar informacion
        print("Información del Profesor:")
        print(f"ID: {self.informacion['id']}")
        print(f"Nombre: {self.informacion['nombre']}")
        print(f"Email: {self.informacion['email']}")
        print(f"Especialización: {self.informacion['especializacion']}")
    
    def __str__(self):
        return f"ID: {self.informacion['id']}\nNombre: {self.informacion['nombre']}\nEmail: {self.informacion['email']}\nEspecializacion: {self.informacion['especializacion']}"
    
class Alumno (Personal_Universitario):#Clase heredera de superclase
    def __init__ (self, id, nombre, email, creditos_aprobados):#Añadimos el dato creditos aprobados
        super().__init__(id, nombre, email)
        self.informacion["creditos_aprobados"] = creditos_aprobados
    
    def mostrar_informacion(self):#Metodo para mostrar informacion
        print("Información del Alumno:")
        print(f"ID: {self.informacion['id']}")
        print(f"Nombre: {self.informacion['nombre']}")
        print(f"Email: {self.informacion['email']}")
        print(f"Créditos Aprobados: {self.informacion['creditos_aprobados']}")
    
    def __str__(self):
        return f"ID: {self.informacion['id']}\nNombre: {self.informacion['nombre']}\nEmail: {self.informacion['email']}\nCreditos Aprobados: {self.informacion['creditos_aprobados']}"
    
#Creamos objetos para probar la herencia y para mostrar informacion  
oficina = Oficina(id=1, nombre="Juan", email="juan@example.com", puesto="Secretario")
oficina.mostrar_informacion()
print("\n")
profesor = Profesor(id=2, nombre="Maria", email="maria@example.com", especializacion="Matemáticas")
profesor.mostrar_informacion()
print("\n")
alumno = Alumno(id=3, nombre="Pedro", email="pedro@example.com", creditos_aprobados=120)
alumno.mostrar_informacion()
```

**3) Escoge la temática que prefieras y define una estructura de clases y subclases donde practiques la herencia en al menos dos niveles (herencia multi nivel). Crea una estructura de mínimo 5 clases, con sus respectivos constructores y métodos. Crea algunos objetos y comprueba que la herencia este funcionando. 
Si puedes, incluye alguna técnica de polimorfismo.**

```python
class Vehiculo:#Superclase 
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def conducir(self):
        return f"Conduciendo el vehículo {self.marca} {self.modelo}"
    
    def __srt__(self):
        print ("Estas en vehiculo")

class Vehiculo_terrestre(Vehiculo):#Clase heredera de superclase
    def __init__(self, marca, modelo, tipo_combustible):
        super().__init__(marca, modelo)
        self.tipo_combustible = tipo_combustible

    def conducir(self):
        return f"Conduciendo el vehículo terrestre {self.marca} {self.modelo} con {self.tipo_combustible}"
    
    def __srt__(self):
        print ("Estas en vehiculo terrestre")

class Vehiculo_acuatico(Vehiculo):#Clase heredera de superclase
    def __init__(self, marca, modelo, eslora):
        super().__init__(marca, modelo)
        self.eslora = eslora

    def navegar(self):
        return f"Navegando con el vehículo acuático {self.marca} {self.modelo} de eslora {self.eslora} metros"
    
    def __srt__(self):
        print ("Estas en vehiculo acuatico")

class Vehiculo_aereo(Vehiculo):#Clase heredera de superclase
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo)
        self.autonomia = autonomia

    def volar(self):
        return f"Volando con el vehículo aéreo {self.marca} {self.modelo} con autonomía de {self.autonomia} km"
    
    def __srt__(self):
        print ("Estas en vehiculo aereo")

class Coche(Vehiculo_terrestre):#Clase heredera de segundo nivel
    def __init__(self, marca, modelo, tipo_combustible, num_puertas):
        super().__init__(marca, modelo, tipo_combustible)
        self.num_puertas = num_puertas

    def conducir(self):
        return f"Conduciendo el coche {self.marca} {self.modelo} de {self.num_puertas} puertas"
    
    def __srt__(self):
        print ("Estas en un Coche")

class Avion(Vehiculo_aereo):#Clase heredera de segundo nivel
    def __init__(self, marca, modelo, autonomia, num_asientos):
        super().__init__(marca, modelo, autonomia)
        self.num_asientos = num_asientos

    def volar(self):
        return f"Volando con el avión {self.marca} {self.modelo} con autonomía de {self.autonomia} km y {self.num_asientos} asientos"
    
    def __srt__(self):
        print ("Estas en un Avion")
        
class Barco(Vehiculo_acuatico):#Clase heredera de segundo nivel
    def __init__(self, marca, modelo, eslora, tipo_barco):
        super().__init__(marca, modelo, eslora)
        self.tipo_barco = tipo_barco

    def zarpar(self):
        return f"Zarpando con el barco {self.marca} {self.modelo} con una eslora de {self.eslora} m y {self.tipo_barco} es su clasificacion"

    
    def __srt__(self):
        print ("Estas en un Barco")
        
#Creamos objetos para probar la herencia y el polimorfismo
coche = Coche(marca="Toyota", modelo="Corolla", tipo_combustible="gasolina", num_puertas=4)
avion = Avion(marca="Boeing", modelo="737", autonomia=3000, num_asientos=150)
barco = Barco(marca="Beneteau", modelo="Oceanis Clipper 523", eslora=17, tipo_barco="Velero")

print(coche.conducir())
print("\n")
print(avion.volar())
print("\n")
print(barco.zarpar())
print("\n")
# Polimorfismo: usando el método de la clase base en objetos de subclases
vehiculo_terrestre = Vehiculo_terrestre(marca="Ford", modelo="Focus", tipo_combustible="diesel")
print(vehiculo_terrestre.conducir())# Método específico de Vehiculo_terrestre
print("\n")
vehiculo_acuatico = Vehiculo_acuatico(marca="Yamaha", modelo="WaveRunner", eslora=3)
print(vehiculo_acuatico.conducir())  # El método conducir es heredado de la clase base
print(vehiculo_acuatico.navegar())   # Método específico de Vehiculo_acuatico
print("\n")
vehiculo_aereo = Vehiculo_aereo (marca="Mc Donnell Douglas", modelo="EF-18 Horne", autonomia="3700")
print(vehiculo_aereo.conducir())  # El método conducir es heredado de la clase base
print(vehiculo_aereo.volar())   # Método específico de Vehiculo_aereo
```
